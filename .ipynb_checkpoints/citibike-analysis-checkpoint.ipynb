```python
per_day_member = trip_df.groupby(by=['member_casual', 'start_date'])['ride_id'].count()

per_day_member.head()
```

```python
per_day_member = per_day_member.reset_index()
per_day_member = per_day_member.rename(columns={'ride_id': 'count'})
per_day_member["start_date"] = pd.to_datetime(per_day_member['start_date'].astype(str))

per_day_member.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

sns.lineplot(data=per_day_member, 
             x="start_date", 
             y="count", 
             hue='member_casual', 
             ax=ax)
```

```python
sns.relplot(data=per_day_member, 
            x="start_date", 
            y="count",
            col="member_casual", 
            hue="member_casual",
            kind="line", 
            height=8)
```

# Mapping

```python
trip_gdf = gpd.GeoDataFrame(trip_df, crs=4326,
                           geometry=gpd.points_from_xy(trip_df.start_lng, 
                                                       trip_df.start_lat))

trip_gdf.head()
```

```python
trip_gdf.crs
```

```python
trip_gdf.plot()
```

```python
trip_gdf = trip_gdf.to_crs(2263)

trip_gdf.crs
```

```python
trip_gdf.head()
```

```python
nycbor = gpd.read_file('borough-boundaries')

nycbor.head()
```

```python
nycbor.plot()
```

```python
nycbor.crs
```

```python
nycbor = nycbor.to_crs(2263)

nycbor.crs
```

```python
# spatial join nta data to the school points data with operation = 'intersects'
trips_merged = gpd.sjoin(trip_gdf, 
                         nycbor, 
                         how="inner", 
                         op='intersects')

trips_merged.head()
```

```python
trips_merged['boro_code'].value_counts()
```

```python
boro_counts = trips_merged.groupby(by='boro_code')['ride_id'].count()
boro_counts = boro_counts.rename('trip_counts')

boro_counts
```

```python
boro_merged = nycbor.merge(boro_counts, 
                           left_on='boro_code', 
                           right_on='boro_code', 
                           how='left')

boro_merged
```

```python
boro_merged['count_per_area'] = boro_merged['trip_counts'] / (boro_merged['shape_area'] / 27878400)

boro_merged.head()
```

```python
boro_merged.plot(column='count_per_area')
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

boro_merged.plot(column='count_per_area', 
                 legend=True, 
                 ax=ax)

nycbor.plot(ax=ax, 
            zorder=0, 
            edgecolor='black', 
            color='None')
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

boro_merged.plot(column='count_per_area', 
                 legend=True,
                 scheme='NaturalBreaks', 
                 k=7,
                 ax=ax)

nycbor.plot(ax=ax, 
             zorder=0, 
             edgecolor='black', 
             color='None')
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

boro_merged.plot(column='count_per_area', 
                 legend=True,
                 k=7,
                 scheme='NaturalBreaks', 
                 ax=ax)

nycbor.plot(ax=ax, 
             zorder=0, 
             edgecolor='black', 
             color='None')

plt.axis('off')
```
